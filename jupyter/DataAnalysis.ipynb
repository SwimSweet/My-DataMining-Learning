```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Salaries_data=pd.read_csv(r'C:\Users\huang\Desktop\lahman\Salaries.csv')
```

```python
grouped=Salaries_data.groupby(['teamID','yearID'])
```

```python
New_Salaries=grouped['salary'].sum()
```

```python
Team_data=pd.read_csv(r'C:\Users\huang\Desktop\lahman\Teams.csv')
```

```python
New_Salaries1=pd.DataFrame(New_Salaries)
```

```python
New_merge_data=pd.merge(New_Salaries1,Team_data,left_index=True,right_on=['teamID','yearID'])
```

```python
New_merge_data_split=New_merge_data[['salary','yearID','teamID','W']]
```

```python
New_merge_data_split[New_merge_data_split['yearID']==1997].head()
```

```python
fig,axes=plt.subplots(6,3,sharex=True,sharey=True)
```

```python
fig,axes=plt.subplots(6,3,figsize=(12, 15),sharex=True,sharey=True)
for i in range(6):
    for j in range(3):
        New_merge_data_split1=New_merge_data_split[New_merge_data_split['yearID']==1997+3*i+j]
        idx_1=New_merge_data_split1['teamID']=='OAK'
        idx_2=New_merge_data_split1['teamID']!='OAK'
        axes[i,j].scatter(New_merge_data_split1[idx_1]['salary'],New_merge_data_split1[idx_1]['W'],color='red',label='OAK',marker = 'x')
        axes[i,j].scatter(New_merge_data_split1[idx_2]['salary'],New_merge_data_split1[idx_2]['W'],color='blue',label='Other',marker = 'o')
        axes[i,j].set_title(str(1997+3*i+j)+"year")
        axes[i,j].grid(True)
plt.subplots_adjust()
plt.legend(loc = 'upper right')  
```
