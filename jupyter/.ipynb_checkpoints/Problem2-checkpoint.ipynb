```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
countries=pd.read_csv(r'F:\数据挖掘学习\dataSet\2014_data\countries.csv')
```

```python
income=pd.read_excel(r'F:\数据挖掘学习\dataSet\2014_data\indicator gapminder gdp_per_capita_ppp.xlsx',index_col=0)
```

```python
income=income.dropna()
```

```python
fig,ax=plt.subplots(figsize=(20, 10))
ind=np.arange(1,len(income)+1)
rect=ax.bar(ind,height=np.array(income[2000]))
for rec in rect:
    x=rec.get_x()
    height=rec.get_height()
    ax.text(x+0.1,1.02*height,str(height))
ax.set_xticks(ind)
ax.set_xticklabels(income.index)
ax.set_ylabel("income per person ")
ax.set_xlabel("countries")
ax.grid(True)
plt.show()
```
